# 설문조사 항목 (input)



## 칼로리 조정 항목 (강아지의 신체 상태에 따른 칼로리 조절)
- 종 
- 연령
- 성별
- 중성화 여부
- 견종 
- 무게
- 체중 증가, 감소중 원하는 방향 질문
- 활동성

## 영양소 조정 항목  (강아지의 병명에 따른 영양소 조절)
- 원하는 개선 부위
- 알러지 여부
- 병력, 현재 가지고 있는 질병


## 급여량 조정 항목  (급여 주기, 급여 형태 결정)
- 현재 먹고 있는 강아지의 사료 형태
- 원하는 사료 급여 주기

## 식품 조정 항목
- 희망 재료, 혹은 사용자 본인이 소유하고 있는 재료
- 재료의 무게 (선택 항목)
- 단백질, 지방, 탄수화물의 비율
